## Manually Face detection

```python
import cv2
import matplotlib.pyplot as plt

img=cv2.imread("men.jpg")
img=cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
plt.imshow(img)
plt.grid()
```

```python
cv2.rectangle(img,(226,150),(310,40),(255,0,0),4)
plt.imshow(img)
```

## Face detection using harcascade

```python
face="/Users/jeetbiswas/Desktop/Internship/haarcascade_frontalface_default.xml"
img=cv2.imread("men.jpg")
img=cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
# img=cv2.imread(img)
model=cv2.CascadeClassifier(face)

facepoints=model.detectMultiScale(img)
for x,y,w,h in facepoints:
    cv2.rectangle(img,(x,y),(x+w,y+h),(255,0,0),4)
    plt.imshow(img)
```
