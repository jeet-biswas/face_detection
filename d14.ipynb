```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
image=cv2.imread('dog.jpg')
image.shape
image
```

```python
# computer vision is a field of ai that enable machine to derives meaningful information from digital images,videos and other visual inputs.

# it is a field of ai that enable machiens to see, understand and interpret visual information -just like human do 
```

```python
#  For creating colors we use depth stack-dstack
```

```python
img=cv2.cvtColor(image,cv2.COLOR_BGR2RGB)
```

```python
# Gray image
gray_image=cv2.cvtColor(image,cv2.COLOR_BGR2GRAY)
```

```python
plt.imshow(img)
plt.axis('off')
```

```python
plt.imshow(gray_image)
```

```python
b=np.random.randint(0,255,(10,10,4))
```

```python
plt.imshow(b,alpha=0.9) # alpha value decide how deep the color is range is (0-1) 0 means less and 1 means deep
```

```python
r=np.full((50,50),69)
g=np.full((50,50),0)
b=np.full((50,50),0)
```

```python
rgb=np.dstack((r,g,b))
plt.imshow(rgb)
# plt.axis('off')
```

```python
#  Blur image 
blur=cv2.GaussianBlur(image,(99,99),5)
plt.imshow(blur)
```

```python
# capturing image through webcam
cap=cv2.VideoCapture(0)
if cap.isOpened():
    ret,frame=cap.read()
    print(ret)
    print(frame)
else:
    ret=False

my_image=cv2.cvtColor(frame,cv2.COLOR_BGR2RGB)
plt.imshow(my_image)
plt.title("My Photo")
plt.axis("off")
cap.release()
```
